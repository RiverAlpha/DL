3. 给定方阵 $A$，$A + A^\top$ 总是对称矩阵吗？

是的。 判断一个矩阵 $M$ 是否对称，只需看 $M^\top = M$ 是否成立。利用前两题的结论：令 $M = A + A^\top$，则 $M^\top = (A + A^\top)^\top$。根据习题 2 分配转置：$M^\top = A^\top + (A^\top)^\top$。根据习题 1 化简：$M^\top = A^\top + A = M$。得证。

4 & 5. 形状为 (2, 3, 4) 的张量 X，len(X) 的输出是什么？

讲解：在 Python 中，len() 函数总是返回张量的**第 0 个轴（axis=0，最外层维度）**的长度。无论张量有多少维，len() 只看第一个数字。因此形状 (2, 3, 4) 的 len(X) 输出是 2。

6. 运行 A / A.sum(axis=1) 会发生什么？

讲解：会报错（或者得到错误的结果）。

假设 $A$ 是一个形状为 (5, 4) 的矩阵。A.sum(axis=1) 是按行求和，它会把第 1 轴“压扁”，返回一个形状为 (5,) 的一维张量。

当你执行除法时，PyTorch 会触发广播机制（Broadcasting）。

广播规则要求从最右边（后向维度）开始对齐：A 的尾部是 4，而求和结果的尾部是 5。

因为 $4 \neq 5$，所以无法广播计算。正确做法是保持维度数量不变（Keepdims）：A / A.sum(axis=1, keepdims=True)，这样除数形状就变成了 (5, 1)，可以完美广播。

9. linalg.norm 作用于 3 维以上张量计算的是什么？讲解：它默认计算的是所有元素的平方和的平方根。本质上，它把多维张量“展平”成了一个长长的一维向量，然后计算 $L_2$ 范数。对于矩阵来说，这被称为 Frobenius 范数。

10. 计算大矩阵连乘 $(AB)C$ 和 $A(BC)$ 的内存与速度差异

讲解：矩阵乘法 $M_{m \times n} \times N_{n \times p}$ 的乘法运算量是 $O(mnp)$。结合律决定了结果一致，但运算量天差地别：计算 $(AB)C$：先算 $AB$，运算量 $2^{10} \cdot 2^{16} \cdot 2^5 = 2^{31}$。生成大小为 $2^{10} \times 2^5$ 的中间矩阵（极小）。再算乘 $C$，运算量 $2^{10} \cdot 2^5 \cdot 2^{14} = 2^{29}$。总运算量大约是 $2^{31}$ 级别。计算 $A(BC)$：先算 $BC$，运算量 $2^{16} \cdot 2^5 \cdot 2^{14} = 2^{35}$。生成大小为 $2^{16} \times 2^{14}$ 的巨大中间矩阵。再算乘 $A$，运算量 $2^{10} \cdot 2^{16} \cdot 2^{14} = 2^{40}$。总运算量大约是 $2^{40}$ 级别。结论：$(AB)C$ 的速度要快几百倍，且内存占用极小。这被称为“矩阵链乘法”优化。

11. 计算 $AB$ 与 $AC^\top$ 的速度差异及内存克隆问题

讲解：虽然 $B$ 和 $C^\top$ 形状相同，但在计算机内存中，标准的多维数组是**按行优先（Row-major，C-contiguous）**存储的。计算矩阵乘法时，我们需要取左矩阵的行和右矩阵的列做内积。取 $B$ 的列在内存中是跳跃读取的，缓存命中率低；而 $C^\top$ 的“列”其实就是原始矩阵 $C$ 的“行”，内存中是连续的，读取极快。因此 $AC^\top$ 计算速度通常快于 $AB$。如果不克隆内存直接初始化 $C = B^\top$：此时 $C$ 只是 $B$ 的一个视图（View），改变了步长（Stride）的读取方式，内存并未重新排列。所以计算此时的 $AC^\top$ 就相当于原本的 $AB$，丧失了内存连续性的速度优势。

12. 堆叠三个 $100 \times 200$ 矩阵并还原讲解：如果将 $A, B, C$ 沿着“第三个轴”（即 dim=2）堆叠，张量的形状将变为 (100, 200, 3)。要取出 $B$，只需要对最后一个维度的索引 1（第二个坐标）进行切片。

```python
A, B, C = torch.randn(100, 200), torch.randn(100, 200), torch.randn(100, 200)
# 沿着 dim=2 (第三个轴) 堆叠
X = torch.stack([A, B, C], dim=2)
print(X.shape)  # 输出: torch.Size([100, 200, 3])

# 提取第二个坐标（索引为1）来恢复 B
B_recovered = X[:, :, 1]
print(torch.equal(B, B_recovered))  # 输出: True
```
